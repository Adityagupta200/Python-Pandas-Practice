```python
import pandas as pd
df = pd.read_csv("data/weather_by_cities.csv", parse_dates=["day"])
df.set_index('day', inplace=True)
df
```

```python
new_df = df.fillna({
    "temperature" : 0,
    "windspeed": 0,
    "event": "no event"
})
new_df
```

```python
new_df_with_ffil = df.fillna(method="ffill")
new_df_with_ffil
```

```python
newdf_with_ffil_with_lim = df.fillna(method="ffill", limit=1)
newdf_with_ffil_with_lim
```

```python
new_df_with_interpolation = df.interpolate(method="time")
new_df_with_interpolation
```

```python
new_df_with_all_na_values = df.dropna(thresh=2)
new_df_with_all_na_values
```

```python
dt = pd.date_range("01-01-2017", "01-12-2017")
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```

# Dropping duplicates in pandas

```python
duplicate_df = pd.DataFrame(
    {"Names": ["Raj", "Rajat", "Gaurav", "Karan", "Aasif", "Raj"],
     "Marks": [56, 89, 90, 75, 24,56],
     "City": ["New Delhi", "Pune", "Banglore", "Mumbai", "Kolkata", "New Delhi"]
    }
)
duplicate_df
```

```python
duplicate_df.drop_duplicates(subset="Names",keep="first")
```
