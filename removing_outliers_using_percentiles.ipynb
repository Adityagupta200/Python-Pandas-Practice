```python
import pandas as pd

df = pd.read_csv("data/heights.csv")
max_threshold = df["height"].quantile(0.95)
min_threshold = df["height"].quantile(0.05)

outlier_removed_df_using_percentile_method = df[(df["height"] < max_threshold) & (df["height"] > min_threshold)]
print(f"Number of outliers removed: {df.shape[0] - outlier_removed_df_using_percentile_method.shape[0]}")
```

```python
min_threshold, max_threshold = df_property["price_per_sqft"].quantile([0.001, 0.999])
min_threshold, max_threshold
```

```python
df_property[df_property["price_per_sqft"] < min_threshold]
df_property[df_property["price_per_sqft"] > max_threshold]
```

```python
outlier_removed_df_using_percentile_method_on_big_dataset = df_property[(df_property["price_per_sqft"] > min_threshold) & (df_property["price_per_sqft"] < max_threshold)]

number_of_outliers_removed = df_property.shape[0] - outlier_removed_df_using_percentile_method_on_big_dataset.shape[0]
print(f"Number of outliers removed: {number_of_outliers_removed}")
```
